```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
```

```python
df=pd.read_csv('/content/drive/MyDrive/Python/Dataset/Used_Bikes.csv')
df
```

```python
df['brand'].unique()
```

```python
df['owner'].unique()
```

```python
new_df=df.drop(['bike_name','city'], axis='columns')
new_df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
new_df['owner']=le.fit_transform(new_df['owner'])
```

```python
class_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(class_mapping)
```

```python
new_df['brand']=le.fit_transform(new_df['brand'])
```

```python
class_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(class_mapping)
```

```python
new_df
```

```python
x=new_df.drop('price',axis='columns')
y=new_df['price']
```

```python
new_df.isna().sum()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
print(type(x_train))
```

```python
print(type(x_test))
```

```python
model=SVR()
model.fit(x_train,y_train)
```

```python
x_train
```
