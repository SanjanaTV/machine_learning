```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/content/drive/MyDrive/Python/Dataset/hiring.csv')
df
```

```python
new_df=df.drop('experience',axis='columns')
new_df
```

```python
new_df['test_score']=new_df['test_score'].fillna(new_df['test_score'].median())
new_df
```

```python
x=new_df.drop('salary',axis='columns')
y=new_df['salary']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
model=RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
x_train
```

```python
x_test
```

```python
y_predicted=model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
