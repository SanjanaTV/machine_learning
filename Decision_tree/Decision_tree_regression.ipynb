```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/Python/Dataset/homeprices (1).csv")
df
```

```python
%matplotlib inline
plt.xlabel('axis')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
new_df=df.drop('price',axis='columns')
new_df
```

```python
price=df.price
price
```

```python
reg=DecisionTreeRegressor()
reg.fit(new_df,price)
```

```python
reg.predict([[3300]])
```
